## Les bases

```python
import numpy as np

data = np.zeros((6, 6)) # creates an array of shape (6,6) filled with the value 0
print(data)
print(type(data))
print(data.shape)
print(data.dtype)
```

```python
data[5][3] = 36
data[4,2] = 42
data[2] = 2

print(data)
```

```python
data[2] = data[2] * 2
print(data)
```

```python
print("Third line:")
print(data[2])

print("Five first lines (from 0 (included) to 5 (non inclusive))")
print(data[0:5])

print("Third column")
print(data[:,2])

print("Advanced slice")
print(data[4:6,2:4])
```

## Reshape

```python
print(data)
reshaped = data.reshape(3,12)
print(reshaped)
```

## Opérations mathématiques

```python
# New array for the following examples

my_array = [
	[1, 2, 3, 4, 500],
	[10, 9, 8, 7, 6],
	[0, 100, 1, 0, -5]
]
array = np.array(my_array)
print(array)
```

### Arithmétique

```python
array = 2 * array
print(array)

print()

array[1] *= 10
print(array)
```

```python
print("Sum of all elements: ", array.sum())
```

### Analyse

```python
print("Minimum: ", array.min())
print("Maximum: ", array.max())
```

```python
print("index of the maximum value (in a flatten array)  :", array.argmax())
print("index of the maximum value (by line):            :", array.argmax(0))
print("index of the maximum value (by column)           :", array.argmax(1))
```

```python
print(array.mean())
print(array.std())
```

```python
np.sign(array)
```

### Fichiers npy et npz

Formats de fichiers spécifiques à numpy :
- **npy** : Stocker un seul array
- **npz** : Stocker plusieurs arrays

#### Écriture

```python
x = dataset[:,0:2]
y = dataset[:,-1]
print(x)
print(y)

np.save("dataset/x.npy", x)
np.save("dataset/y.npy", y)

np.savez("dataset/dataset.npz", x=x, y=y)
```

#### Lecture

```python
loaded_x = np.load("dataset/x.npy")

print(loaded_x)
```

```python
loaded_dataset = np.load("dataset/dataset.npz")
print(loaded_dataset)

loaded_y = loaded_dataset["y"]
print(loaded_y)
```
