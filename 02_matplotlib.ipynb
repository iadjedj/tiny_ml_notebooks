## Import

Par convention on l'importera sous le nom `plt`

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Création de courbes et de droites

```python
# Create a "dataset" with y = 3*x^2 + 2
x = np.linspace(-2, 2, 20)
y = 2 * x**3 + 2
```

### `plt.plot()` Courbe

```python
plt.plot(x, y)
```

### `plt.scatter()` nuage de points

```python
plt.scatter(x, y)
```

## Paramètres de base

```python
plt.plot(x, y, c='red')            # c='r' also works
plt.scatter(x, y, s=100, c='y')    # s: size of the dots
plt.xlabel("the x title")
plt.ylabel("the y title")
plt.title("The title")
plt.show() # Not mandatory with Jupyter but removes the "print"
```

# Figures

```python
plt.figure(figsize=(6, 2)) # 6x2 inches (yes, inches !!)
plt.plot(x, y)
plt.title("A nice figure")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, sharey=True)
fig.set_size_inches(6, 2)

axes[0].set_title("plot()")
axes[0].plot(x, y)

axes[1].set_title("scatter() of y*2")
axes[1].scatter(x, y * 2)

plt.show()
```

```python
fig, axes = plt.subplots(2, 2)
fig.set_size_inches(12, 6)
fig.set

axes[0, 0].set_title("plot()")
axes[0, 0].plot(x, y)

axes[0, 1].set_title("plot() in red")
axes[0, 1].plot(x, y, c='r')

axes[1, 0].set_title("scatter()")
axes[1, 0].scatter(x, y)

axes[1, 1].set_title("scatter() with big stars instead of points\nwhere color depends on y and a title in red", c='r')
axes[1, 1].scatter(x, y, c=y, marker='*', s=400)

plt.tight_layout() # automatically adjusts the margins
plt.show()
```

# Exemple : Iris dataset

Combiner `plt.scatter()` avec une couleur qui varie peut permettre de visualiser un problème de classification.

Voici un exemple avec un célèbre dataset de demonstration, le dataset **Iris** (1936)




L'objectif est de reconnaitre trois types de fleurs (des Iris) en fonction de la taille de leurs pétales et sépales

```python
# Import the iris dataset using sklearn (next chapter)
%pip install -q scikit-learn
from sklearn.datasets import load_iris

dataset = load_iris()
x_iris = dataset.data
y_iris = dataset.target
target_names = dataset.target_names
feature_names = dataset.feature_names
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

sc = ax[0].scatter(x_iris[:, 0], x_iris[:, 1], c=y_iris)
ax[0].set_xlabel(dataset.feature_names[0])
ax[0].set_ylabel(dataset.feature_names[1])

ax[1].scatter(x_iris[:, 2], x_iris[:, 3], c=y_iris)
ax[1].set_xlabel(dataset.feature_names[2])
ax[1].set_ylabel(dataset.feature_names[3])

handles, labels = sc.legend_elements()
ax[1].legend(handles=handles, labels=list(dataset.target_names), loc='lower right')

plt.tight_layout()
plt.show()
```
