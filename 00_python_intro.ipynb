## Les boucles

Les boucles `while` et `for` existent en Python :
### `while`

```python
cities = ["Lyon", "Paris", "Rennes", "Annecy", "New York"]

i = 0
while i < len(cities):
	print(f"- {cities[i]}")
	i += 1
```

### `for`

```python
for idx in range(len(cities)):
	print(f"- {cities[idx]}")
```

### À la Python
On préférera des syntaxes plus "pythoniques"

```python
for city in cities:
	print(f"- {city}")
```

## Les conditions

```python
name = "Pierre"
age = 42

if age < 6:
	print("Baby shark!")
elif 6 <= age < 18: # 6 repeated here to demonstrate syntax min < x < max
	print("You are too young to party here!")
elif name == your_teacher["name"] and age == your_teacher["age"]:
	print("Have we met before?")
elif age > 60:
	print("Welcome old man!")
else:
	print("Welcome here!")
```

## Les fonctions

```python
def say_hi(name, age=None):
	print("Hello " + name) # For simple strings, instead of f-string a simple concatenation is possible
	if age is not None:
		print(f"You are {age} years old!")

say_hi("Ilann", 34)
print()
say_hi("Pierre")
```
